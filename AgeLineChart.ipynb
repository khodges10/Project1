```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
nba_csv = "Data/NBAKeyStats.csv"
nba_key = pd.read_csv(nba_csv)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
nba_key = nba_key.groupby(["Year", "Status"])
nba_average = nba_key["Age"].mean()
nba_average.unstack().plot(ax=ax)
```

```python
plt.title("Age Representation by Team Status")
plt.xlabel("Year")
plt.ylabel("Age")
```

```python
plt.savefig("AgeLineChart.png", bbox_inches="tight")
plt.show()
```
