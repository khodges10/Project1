```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
nba_csv = "Data/NBAKeyStats.csv"
nba_key = pd.read_csv(nba_csv)
nba_key.head()
```

```python
nba_keydoe = nba_key.groupby(["Status"])
nba_barsgraph = nba_keydoe.mean()["Rank"]
```

```python
nba_barsgraph.plot(kind="bar", x="Status", y="League Ranking", grid=False, figsize=(8,8), 
                   title="Net +/- Ranking", color=['black', 'green', 'blue', 'cyan'])
```

```python
plt.show()
plt.savefig("NetRanking.png", bbox_inches="tight")
```
