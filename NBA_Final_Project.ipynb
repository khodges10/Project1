```python
#Age line graph
fig, ax = plt.subplots(figsize=(15,7))
nba_key = nba_key.groupby(["Year", "Status"])
nba_average = nba_key["Age"].mean()
nba_average.unstack().plot(ax=ax)

#labels

plt.title("Age Representation by Team Status")
plt.xlabel("Year")
plt.ylabel("Age")
plt.savefig("AgeLineChart.png", bbox_inches="tight")
```

```python
# formula for Power ranking
nba_key["PR"] = ((nba_key["SRS"] + nba_key["MOV"]) + (nba_key["ORtg"] - 
                        nba_key["DRtg"]*1.25)) * (nba_key["eFG%"] * nba_key["TS%"]) + 10
```

```python
#setting index to PR
nba_key = nba_key.sort_values("PR", ascending=False)
nba_key = nba_key.reset_index(drop=True)
nba_key.round(2)
# 10 of top 16 teams were champions
```

```python
nba_keydoe = nba_key.groupby(["Status"])
nba_barsgraph = nba_keydoe.mean()["PR"]
```

```python
#plot bar for Power Rank
nba_barsgraph.plot(kind="bar", x="Status", y="Power Ranking", grid=False, figsize=(8,8), 
                   title="Average Power Ranking by Status", color=['black', 'green', 'blue', 'cyan'])
#save fig
plt.savefig("PRanking.png", bbox_inches="tight")
plt.show()
```
