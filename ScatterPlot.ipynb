```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
nba_csv = "Data/NBAKeyStats.csv"
nba_key = pd.read_csv(nba_csv)
nba_key.head()
```

```python
nba_key_simple = nba_key[["Year", "Status", "Team", "SRS", "eFG%", "Pace", "3PAr", "FTr", "TS%", 
                          "ORB%", "DRB%", "eFG%.1"]]
nba_key_simple.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(x_vars=["SRS"], y_vars=["eFG%"], data=nba_key_simple, hue="Status", height=5)
```

```python
plt.title("SRS & eFG%")
```

```python
plt.savefig("SRSvsEFGScatter.png", bbox_inches="tight")
plt.show()
```
